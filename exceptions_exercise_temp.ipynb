### Task 01
Write Python code to calculate the square root of a number. The user is:

1.	Prompted for a number
2.	Given the square root of that number
3.	Asked for another number.

The program should keep asking for a number until the user indicates that they have finished data input by entering “end”.

If the user enters anything that is not a valid float number, display an error message of your choosing and prompt them to enter the number again.

Place your code in a separate function called calculate_square_root


```python
# Task 01 - Square root calculation 

def square_root(value1: float) -> float:
    # Define a function to find the square root of a number
    # use the mat library
    import math
    return math.sqrt(value1)
    
    
def test_square_root():
    val1 = 25.0
    val2 = 9.0
    val3 = 12.0
    print("The square root of", val1, "is:", square_root(val1))
    print("The square root of", val2, "is:", square_root(val2))
    print("The square root of", val3, "is:", square_root(val3))


def calculate_square_root():
    # init variables
    user_input = 'end'
    sqroot = 0
    
    while True:
        try:
            # user input
            entry = "Enter a number or 'end' to exit: "
            user_input = input(entry)
            # allow for an exit
            if user_input == 'end':
                break
            # calculate the square root
            sqroot = square_root(float(user_input))
            # print the result
            print('The square root of {0} is {1}'.format(user_input, sqroot))
        except Exception as err:
            print("Houston we have a problem!", err)
        
    
def test_calculate_square_root():
    calculate_square_root()
    
    
# Run all of your test functions from the following 'if' statement
if __name__ == '__main__':
    test_calculate_square_root()
    
```
